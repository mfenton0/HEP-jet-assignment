```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
bins_w = 0.01
bins = np.arange(-4, 4, bins_w)
```

```python
def gau(x, gau_loc1, gau_scale1):
    gau_y1 = (x-gau_loc1)/gau_scale1
    return np.exp(-gau_y1**2/2)/np.sqrt(2*np.pi)/gau_scale1
```

```python
plt.figure(figsize=(30, 20))
dic_popt = {}
dic_perr = {}
scales = [4, 5, 6, 7, 8]
for i, scale in enumerate(scales):
    num = 10**scale
    plt.subplot(2, 3, i+1)
    target = np.random.normal(size=num)
    hist, _ = np.histogram(target, bins=bins, density=True)
    plt.bar(bins[:-1]+bins_w/2, hist)
    popt, pcov = optimize.curve_fit(gau, bins[:-1]+bins_w/2, hist)
    perr = np.sqrt(np.diag(pcov))
    dic_popt[scale] = popt
    dic_perr[scale] = perr
plt.show()
plt.errorbar(scales, [dic_popt[scale][0] for scale in [4, 5, 6, 7, 8]], yerr = [dic_perr[scale][0] for scale in [4, 5, 6, 7, 8]])
plt.show()
plt.errorbar(scales, [dic_popt[scale][1] for scale in [4, 5, 6, 7, 8]], yerr = [dic_perr[scale][1] for scale in [4, 5, 6, 7, 8]])
plt.show()
```
